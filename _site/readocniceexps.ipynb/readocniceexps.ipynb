### Read in the ice history for the control and ice albedo experiments

aice = ice fraction (0-1)

```python
path_ctl_ice='/glade/scratch/kpegion/archive/gcontrol/ice/hist/'
path_gsnowalb_ice='/glade/scratch/kpegion/archive/gsnowalb/ice/hist/'
files_ctl_ice='gcontrol.cice.h.0001-*.nc'
files_gsnowalb_ice='gsnowalb.cice.h.0001-*.nc'
```

```python
ds_ctl_ice=xr.open_mfdataset(path_ctl_ice+\
                             files_ctl_ice,
                             combine='nested',
                             concat_dim='time')
ds_ctl_ice
```

```python
ds_alb_ice=xr.open_mfdataset(path_gsnowalb_ice+\
                             files_gsnowalb_ice,
                             combine='nested',
                             concat_dim='time')
ds_alb_ice
```

### Calculate the diffence in annual mean ice fraction in albedo exp minus control exp

```python
diff31=ds_alb_ice['aice'].mean(dim='time')-\
       ds_ctl_ice['aice'].mean(dim='time')
```

### Plot the result

```python
plt.contourf(diff31)
plt.colorbar()
```

### Read in wind stress experiment ocean data
